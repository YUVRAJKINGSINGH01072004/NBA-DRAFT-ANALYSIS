```python
df.dropna(subset=['position'], inplace=True)
df.dropna(subset=['height'], inplace=True)
df.dropna(subset=['weight'], inplace=True)
df.dropna(subset=['wingspan'], inplace=True)
df.dropna(subset=['reach_standing'], inplace=True)
```

```python
df.isnull().sum()
```

```python
#data imputation
```

```python
mode_value = df['body_fat_pct'].mode()
df['body_fat_pct'].fillna(mode_value, inplace=True)
```

```python
df.isnull().sum()
```

```python
mean_valuesv = df['standing_vertical'].mean()
df['standing_vertical'].fillna(mean_valuesv, inplace=True)
```

```python
mean_valuemv = df['max_vertical'].mean()
df['max_vertical'].fillna(mean_valuemv, inplace=True)
```

```python
mean_valuebr = df['bench_reps'].mean()
df['bench_reps'].fillna(mean_valuebr, inplace=True)
```

```python

mean_valuetl = df['timeLaneAgility'].mean()
df['timeLaneAgility'].fillna(mean_valuetl, inplace=True)
```

```python
mean_valuett = df['timeThreeQuarterCourtSprint'].mean()
df['timeThreeQuarterCourtSprint'].fillna(mean_valuett, inplace=True)
```

```python
df['player_id'].count()
```

```python
df.drop(columns=['widthHandInches'], inplace = True)
```

```python
df.drop(columns=['timeModifiedLaneAgility'], inplace = True)
```

```python
df.drop(columns=['lengthHandInches'], inplace = True)
```

```python
mean_valuebf = df['body_fat_pct'].mean()
df['body_fat_pct'].fillna(mean_valuebf, inplace=True)
```

```python
#draft sucess
```

```python
avg_bench_reps = df.groupby('drafted')['bench_reps'].mean()
plt.figure(figsize=(10, 6))
avg_bench_reps.plot(kind='bar')
plt.xlabel('Drafted (0 = Not Drafted, 1 = Drafted)')
plt.ylabel('Average Bench Press Reps')
plt.title('Average Bench Press Reps: Drafted vs Not Drafted')

```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='drafted', y='wingspan', data=df)
plt.xlabel('Drafted (1 = Drafted, 0 = Not Drafted)')
plt.ylabel('Wingspan (inches)')
plt.title('Wingspan Distribution for Drafted vs Undrafted Players')

plt.show()

```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='drafted', y='timeLaneAgility', data=df)
plt.xlabel('Drafted (1 = Drafted, 0 = Not Drafted)')
plt.ylabel('Time Lane Agility (seconds)')
plt.title('Time Lane Agility Distribution for Drafted vs Undrafted Players')
```

```python
# position based analysis 
```

```python
max_vertical_per_position = df.groupby('position')['max_vertical'].max().sort_values()
plt.figure(figsize=(10, 6))
max_vertical_per_position.plot(kind='bar')
plt.xlabel('Position')
plt.ylabel('Max Vertical Jump (inches)')
plt.title('Max Vertical Jump Comparison by Position')
```

```python
drafted_players = df[df['drafted'] == True]
position_counts = drafted_players['position'].value_counts()
labels = position_counts.index
sizes = position_counts.values
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels ,autopct=lambda pct: f'{pct:.1f}%\n({int(pct/100.*sum(sizes))})')
plt.title('Distribution of Drafted Players by Position')
```

```python
avg_timeLaneAgility = df.groupby('position')['timeLaneAgility'].mean().sort_values()
plt.figure(figsize=(10, 6))
plt.plot(avg_timeLaneAgility.index, avg_timeLaneAgility.values,marker='o')
plt.xlabel('Position')
plt.ylabel('Average Time Lane Agility (seconds)')
plt.title('Average Time Lane Agility by Position')
plt.show()
```

```python
drafted_players = df[df['drafted'] == True]
avg_height_position = drafted_players.groupby('position')['height'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='position', y='height', data=avg_height_position)
plt.xlabel('Position')
plt.ylabel('Average Height (inches)')
plt.title('Average Height by Position for Drafted Players')

```

```python
avg_jump_height = df.groupby('yearDraft')['bench_reps'].mean().reset_index()
sns.lineplot(x='yearDraft', y='bench_reps', data=avg_jump_height, marker='o')
plt.xlabel('Year Draft')
plt.ylabel('Average Benchreps ')
plt.title('Average Benchreps by Year Draft')
```
